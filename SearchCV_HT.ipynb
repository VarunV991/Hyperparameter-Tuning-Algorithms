```
# Load Libraries
import pandas as pd
import numpy as np
import warnings

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score
from sklearn import datasets

warnings.filterwarnings("ignore")
```

```
# Load Dataset
iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target
```

```
# Initialize classifier
cf = RandomForestClassifier(n_jobs=-1)
```

## GridSearchCV

```
from sklearn.model_selection import GridSearchCV
# Create parameter grid

param_grid = {
    "n_estimators":[100,200,300],
    "criterion":["gini","entropy"],
		"max_features" : ['auto','sqrt'],
		"max_depth" : [1,2,3,5],
		"min_samples_split" : [2,5,10],
		"min_samples_leaf" : [1,2,5],
}
```

```
model = GridSearchCV(estimator=cf,param_grid=param_grid,scoring="accuracy",verbose=10,cv=5,n_jobs=1)
```

```
model.fit(X,y)
```

```
print(f'''Best Score: {model.best_score_} \n Best Parameters: {model.best_params_}''')
```

## RandomizedSearch CV

```
from sklearn.model_selection import RandomizedSearchCV
# Create parameter grid

param_grid = {
    "n_estimators":np.arange(100,1500,100),
    "criterion":["gini","entropy"],
		"max_depth" : np.arange(1,20),
}
```

```
rs_model = RandomizedSearchCV(estimator=cf,param_distributions=param_grid,scoring="accuracy",n_jobs=1,cv=5,verbose=1)
```

```
rs_model.fit(X,y)
```

```
print(f'''Best Score: {rs_model.best_score_} \n Best Parameters: {rs_model.best_params_}''')
```

## With Pipelines

```
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
```

```
# Initialize scaling and decomposition fns
scale = StandardScaler()
pca = PCA()
rf = RandomForestClassifier(n_jobs=-1)
```

```
# Define pipeline

pipe_cf = Pipeline([('scaling',scale),('pca',pca),("rf",rf)])

# Define param grid 

param_grid = {
    "pca__n_components":np.arange(0,3),
    "rf__n_estimators":np.arange(100,1500,100),
    "rf__criterion":["gini","entropy"],
		"rf__max_depth" : np.arange(1,20),
}
```

```
rs_model_pipe = RandomizedSearchCV(estimator=pipe_cf,param_distributions=param_grid,scoring="accuracy",n_jobs=1,cv=5,verbose=1)
```

```
rs_model_pipe.fit(X,y)
```

```
print(f'''Best Score: {rs_model_pipe.best_score_} \n Best Parameters: {rs_model_pipe.best_params_}''')
```
